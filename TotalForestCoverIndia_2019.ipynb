```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import waterfall_chart as wc
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/total%20forest%20cover%20of%20india-2019%20(1)")
df
```

```python
df.rename({'Area(sq km)': 'Area_sq_km'}, axis=1,inplace = True)
df
```

```python
df.drop([6], axis = 0, inplace=True)
df.drop(['Unnamed: 0'], axis=1, inplace= True)
df = df.sort_values('Area_sq_km', ascending=False) 
df
```

```python
plt.figure(figsize=(5,5))                                  
plt.style.use("seaborn")
x = df['Classification']
y = df['Area_sq_km']
width = 0.5
plt.bar(x, y, width, color = ['yellow', 'red', 'green', 'blue', 'cyan', 'fuchsia'])
plt.plot(x, y, color = 'black', marker = 'd')
plt.xticks(rotation=45, ha="right", fontsize=14)
plt.yticks(fontsize = 14)
plt.ticklabel_format(axis="y", style='plain')
plt.title("Forest Cover Area of India for 2019", fontsize=20)
plt.xlabel("Classifacation", fontsize = 16)
plt.ylabel("Area $Km^{2}$", fontsize = 16)
plt.show()
plt.close()
```

```python
wc.plot(x, y, sorted_value = True, net_label='Total Area', rotation_value = 90,
        x_lab='$Classification$', y_lab="Area $Km^{2}$", Title="Forest Cover Area of India for 2019");


plt.title("FOREST COVER AREA OF INDIA FOR 2019", fontsize=18)
plt.xticks(fontsize=14, ha= "right")
plt.yticks(fontsize=14)
plt.ylabel("AREA $Km^{2}$", fontsize = 16)
plt.xlabel("CLASSIFICATION", fontsize = 16)

sns.despine()
plt.show()
```
