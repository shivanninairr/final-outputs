```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/sg-008/Data-of-Indian-Forests/main/Annual%20loss%20of%20primary%20forest%20cover%20and%20tree%20cover.csv")
```

```python
df.head()
```

```python

df.shape
```

```python
df.drop(labels=[0],inplace=True)
```

```python
df.columns = ['Year','Forest_Cover','Tree_Cover']
```

```python
df
```

```python
df_1 = df.iloc[:18,:]
```

```python
df_1
```

```python
df_1.replace(np.nan,0)
```

```python
df_1.replace(np.nan,0)
```

```python

df_1['Year']=df_1['Year'].str.replace(',','')
```

```python
df_1['Tree_Cover']=df_1['Tree_Cover'].str.replace(',','')
```

```python
df_1['Forest_Cover']=df_1['Forest_Cover'].str.replace(',','')
```

```python

df_1['Tree_Cover']=df_1['Tree_Cover'].astype(float)
```

```python

df_1['Forest_Cover']=df_1['Forest_Cover'].astype(float)
```

```python

df_1['Year']=df_1['Year'].astype(int)
```

```python
df_1.dtypes
```

```python
df_1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
s_imputer=SimpleImputer(missing_values=np.nan,strategy='mean',verbose=0)
```

```python
s_imputer=s_imputer.fit(df_1[0:8])
df_1[0:18]=s_imputer.transform(df[0:18])
```

```python
df_1=df_1.tail(11)
df_1
```

```python
df_2=df_1.set_index("Year")
df_2
```

```python
import seaborn
plt.figure(figsize = (8,8))
plt.title("Heat map",fontsize="16")
seaborn.heatmap(df_2, annot = True,fmt= 'g', cmap = 'Greens')
plt.savefig("annual loss heat map.png")
```

```python
X = df_1['Year']
fig, ax = plt.subplots(2, 1, figsize=(9, 7))

z = np.polyfit(X, df_1['Forest_Cover'], 1)
p = np.poly1d(z)
ax[0].plot(X,p(X),"b--")
ax[0].plot(X, df_1['Forest_Cover'], color ='magenta')
ax[0].set_title('Forest Cover', fontsize=14)

z = np.polyfit(X, df_1['Tree_Cover'], 1)
p = np.poly1d(z)
ax[1].plot(X,p(X),"b--")
ax[1].plot(X, df_1['Tree_Cover'], color = 'magenta')
ax[1].set_title('Tree Cover', fontsize=14)

fig.text(0.5, 0.03, 'YEARS', ha='center', fontsize=14)
fig.suptitle('FOREST & TREE COVER USING SIMPLE LINEAR TREND', ha= 'center', fontsize = 18)
fig.text(0, 0.5, 'AREA OF COVER', va='center', rotation='vertical', fontsize=14)
fig.tight_layout(pad=3.0)
plt.savefig("forest and tree cover.png")
```
