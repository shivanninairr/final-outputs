```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/Annual_Loss_Of_Forest_cover_1.csv")
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
df.rename({'Forest_Cover(ha)': 'Forest_Cover', 'Tree_Cover(30%)': 'Tree_Cover'}, axis = 1, inplace = True)
```

```python
df =df.tail(11)
df
```

```python
df_1 =df.set_index("Year")
df_1
```

```python
plt.figure(figsize = (8,8))

seaborn.heatmap(df_1, annot = True,fmt= 'g', cmap = 'viridis_r', linecolor='black', linewidths=0.2)

plt.xticks(fontsize=12)
plt.yticks(fontsize = 12)
plt.title("HEAT MAP FOR ANNUAL LOSS OF TREE AND FOREST COVER :2018",fontsize="16")
plt.ylabel('YEAR', fontsize = 14)
plt.xlabel('In Hectares (Ha)', fontsize = 14)

plt.savefig('Annual_Loss_Tree&Forest_Cover_HEATMAP_2018.png',transparent=False, bbox_inches='tight', pad_inches=0, dpi = 1200)     
plt.show()
plt.close()
```

```python
plt.style.use("seaborn-darkgrid")

X = df['Year']
fig, ax = plt.subplots(2, 1, figsize=(12, 8))

z = np.polyfit(X, df['Forest_Cover'], 1)
p = np.poly1d(z)
ax[0].plot(X,p(X),"crimson")
ax[0].plot(X, df['Forest_Cover'], color ='midnightblue', marker = 'o')
ax[0].set_title('FOREST COVER', fontsize=11)

z = np.polyfit(X, df['Tree_Cover'], 1)
p = np.poly1d(z)
ax[1].plot(X,p(X),"crimson")
ax[1].plot(X, df['Tree_Cover'], color = 'midnightblue', marker = 'o')
ax[1].set_title('TREE COVER', fontsize=11)

fig.text(0.5, 0.03, 'YEARS', ha='center', fontsize=14)
fig.suptitle('LOSS IN FOREST & TREE COVER : 2018', ha= 'center', fontsize = 16)
fig.text(0, 0.5, 'AREA OF COVER IN HECTARES', va='center', rotation='vertical', fontsize=14)
fig.tight_layout(pad=3)

plt.savefig('Loss_in_Forest_Tree_Cover_2018.png',transparent=False, bbox_inches='tight', pad_inches=0, dpi = 1200)     
plt.show()
plt.close()
```

*Interpretation : The above visualisation depicts a clear upward linear trend in loss of both Forest Covers and Tree Covers in the country for years 2008 - 2019. _Trendlines are a visual representation of support and resistance in any time frame. They show direction and speed of a dataset, and also describe patterns in the data showing an increase (upward trend) or decrease (downward trend)_. There has been seen a decrease-increase trend in loss of forest covers from year 2008 to 2013, the peak of the trend line at 2016 was observed with a loss of 30,936 Ha of geographical area. The latest records show that forest cover for 2018 has managed to cope slightly and maintain its forest covers.  

A similar situation has been observed with the Tree Covers of the country. The maximum loss was found to occur in the year 2017 ie. 1,89,677 Ha of the total geographical region whereas 2010 happened to be the most ecological year of the timeline.*
