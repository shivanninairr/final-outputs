```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/2019_forest_cover_statewise_india.csv")
```

```python
df
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df.drop(labels=[38],inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(15,8))                                  
plt.style.use("seaborn")
y = df['2019_Assessment_Total_Forest_Cover']
x = df['State/UT']
width = 0.5
plt.bar(x, y, width, color = 'c')
plt.xticks(x,rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize = 14)
plt.ticklabel_format(axis="y", style='plain')
plt.title("Forest covers-2019", fontsize=20)
plt.xlabel("STATES/UTS", fontsize = 16)
plt.ylabel("FOREST COVER", fontsize = 10)
plt.show()
plt.close()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(211)
plt.plot(df["State/UT"],df["2019_Assessment_VDF"],color='red')
plt.xticks(rotation=45,ha='right')
plt.subplot(212)
plt.plot(df["State/UT"],df["2019_Assessment_MDF"])
plt.xticks(rotation=45,ha='right')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(211)
plt.plot(df["State/UT"],df["scrub"],color='yellow')
plt.xticks(rotation=45,ha='right')
plt.subplot(212)
plt.plot(df["State/UT"],df["2019_Assessment_OF"],color='lime')
plt.xticks(rotation=45,ha='right')
plt.show()
plt.close()
```
