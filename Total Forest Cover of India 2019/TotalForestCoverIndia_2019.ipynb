```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import waterfall_chart as wc
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/total%20forest%20cover%20of%20india-2019%20(1)")
```

```python
df.rename({'Area(sq km)': 'Area_sq_km'}, axis=1,inplace = True)
```

```python
df.drop([6], axis = 0, inplace=True)
df.drop(['Unnamed: 0'], axis=1, inplace= True)
df = df.sort_values('Area_sq_km', ascending= True) 
df
```

```python
plt.figure(figsize=(8,8))                                  
plt.style.use("seaborn")

x = df['Classification']
y = df['Area_sq_km']
width = 0.5

plt.bar(x, y, width, color = ['limegreen','seagreen','yellowgreen','mediumseagreen','olivedrab','tomato'])
plt.plot(x, y, color = 'black', marker = '.')

plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize = 12)
plt.ticklabel_format(axis="y", style='plain')
plt.title("FOREST COVER OF INDIA FOR 2019", fontsize=16)
plt.xlabel("CLASSIFICATION", fontsize = 14)
plt.ylabel("AREA $Km^{2}$", fontsize = 14)

plt.savefig('Total_ForestCover_India_2019.png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close()
```

*Interpretation : It is observed that out of the total geographical area, forest cover has an area about 21.67% which is classified into Moderately dense forest (9.39 %), Open Forest (9.26 %), Very Dense Forest (3.02 %) and Scrubs(1.41%).

```python
wc.plot(x, y, sorted_value = True, net_label='Total Area',x_lab='$Classification$', 
        y_lab="Area $Km^{2}$", Title="Forest Cover Area of India for 2019")

plt.title("FOREST COVER AREA OF INDIA FOR 2019", fontsize=16)
plt.xticks(fontsize=12, ha= "right", rotation = 70)
plt.yticks(fontsize=12)
plt.ylabel("AREA $Km^{2}$", fontsize = 14)
plt.xlabel("CLASSIFICATION", fontsize = 14)

plt.savefig('Total_ForestCover_India_2019_waterfallchart.png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close()
```
