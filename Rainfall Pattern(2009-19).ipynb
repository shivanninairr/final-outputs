```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/Rainfall_pattern(2009-2019)",index_col= False)
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace= True)
df.rename({"JUN(in mm)":"JUN"},axis=1,inplace=True)
df
```

```python
x = df['YEAR']
y = df['JUN-SEP']
plt.figure(figsize=(12,8))
plt.plot(x,y,color='black')
plt.annotate("714.2",(2009.1,714.2),fontsize=15)
plt.annotate("910.7",(2009.8,915),fontsize=15)
plt.annotate("915.4",(2011,915.4),fontsize=15)
plt.annotate("841.7",(2011.7,830),fontsize=15)
plt.annotate("945.8",(2013,945.8),fontsize=15)
plt.annotate("784.2",(2014,784.2),fontsize=15)
plt.annotate("765.4",(2014.7,756),fontsize=15)
plt.annotate("863.7",(2016,863.7),fontsize=15)
plt.annotate("843.7",(2017,843.7),fontsize=15)
plt.annotate("802.4",(2017.7,795),fontsize=15)
plt.annotate("969.4",(2019,969.4),fontsize=15)
plt.ticklabel_format(style='plain')   #Does not convert the axis 
plt.title('YEAR VS RAINFALL(in mm)',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('YEAR',fontsize=18)
plt.ylabel('RAINFALL(in mm)',fontsize=18)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))(np.unique(x)), color = 'red')
plt.grid(True,color="#dddd")
plt.show()
```

```python

x = df['YEAR']
y1 = df['JUN']
y2 = df['JUL']
y3 = df['AUG']
y4 = df['SEP']
width=0.2
  
# plot data in grouped manner of bar type
plt.style.use("seaborn")
plt.figure(figsize=(15,9))
plt.bar(x-0.2, y1, width, color='grey')
plt.bar(x, y2, width, color='magenta')
plt.bar(x+0.2, y3, width, color='pink')
plt.bar(x+0.4, y4, width, color='crimson')
plt.xticks(x,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Teams",fontsize=18)
plt.ylabel("Scores",fontsize=18)
plt.legend(["June", "July", "August","September"],fontsize=16,loc="upper left",ncol=2)
plt.show()
```
