```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/Rainfall_pattern(2009-2019)")
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace= True)
df.rename({"JUN(in mm)":"JUN"},axis=1,inplace=True)
df.head()
```

```python
x = df['YEAR']
y = df['JUN-SEP']

plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(12,8))
plt.plot(x,y,color='midnightblue', marker = 'o', markerfacecolor='teal')

#Adding annotations
plt.annotate("714.2",(2009.1,714.2),fontsize=13)
plt.annotate("910.7",(2009.8,915),fontsize=13)
plt.annotate("915.4",(2011,918.5),fontsize=13)
plt.annotate("841.7",(2011.7,830),fontsize=13)
plt.annotate("945.8",(2013,948.5),fontsize=13)
plt.annotate("784.2",(2014.08,785),fontsize=13)
plt.annotate("765.4",(2014.7,756),fontsize=13)
plt.annotate("863.7",(2016,866.5),fontsize=13)
plt.annotate("843.7",(2017.07,843.7),fontsize=13)
plt.annotate("802.4",(2017.7,793),fontsize=13)
plt.annotate("969.4",(2019,960),fontsize=13)

plt.ticklabel_format(style='plain') 
plt.title('YEAR VS RAINFALL(in mm)',fontsize=16)
plt.xticks(x, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('YEAR',fontsize=14)
plt.ylabel('RAINFALL(in mm)',fontsize=14)

#Polynomial fitting(Trend Line)
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))(np.unique(x)), color = 'crimson')

plt.savefig('Rainfall_Pattern(2009-2019).png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close
```

*Interpretation: The trend line depicts that there has been a very slight increase in the overall rainfall pattern from the year 2009 to 2019 as the slope tends to flatten, whereas there is a significant increase  of 255.2 mm of the rainfall in the year 2019 when compared to 2009.*

```python
x = df['YEAR']
y1 = df['JUN']
y2 = df['JUL']
y3 = df['AUG']
y4 = df['SEP']
width= 0.15
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(15,7))
plt.bar(x-0.15, y1, width, color='orangered')
plt.bar(x, y2, width, color='darkcyan')
plt.bar(x+0.15, y3, width, color='gold')
plt.bar(x+0.3, y4, width, color='darkkhaki')

plt.xticks(x,fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("YEAR",fontsize=14)
plt.ylabel("RAINFALL(in mm)",fontsize=14)
plt.title('YEAR VS RAINFALL(in mm) FOR JUN-SEPT 2009-2019',fontsize=16)
plt.legend(["JUNE", "JULY", "AUGUST","SEPTEMBER"],fontsize=14,loc="upper left",bbox_to_anchor=(1, 1))

plt.savefig('Rainfall_Pattern_JUN-SEPT_(2009-2019).png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close()
```

*Interpretation : As monsoon season approaches the country, month of June experiences the lowest ammount of rainfall whereas July experinces the highest. As the season departs the amount of rainfall in September decreases.*
