```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/Protected_Areas_cleaned.csv")
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace= True)
df.drop([0,1,2], axis=0, inplace= True)
```

```python
Number= df[['Year', 'No. of National Parks','No. of Wild Life Sanctuaries','No. of Community Reserves',
                     'No. of Conservation Reserves','No. of Protected Areas']]
Number.rename({'No. of National Parks' :'National Parks', 'No. of Wild Life Sanctuaries': 'Wild Life Sanctuaries','No. of Community Reserves':'Community Reserves','No. of Conservation Reserves':'Conservation Reserves','No. of Protected Areas':'Total Protected Areas'}, axis= 1, inplace = True)
Number
```

```python
x = Number['Year']
y1 = Number['National Parks']
y2 = Number['Wild Life Sanctuaries']
y3 = Number['Community Reserves']
y4 = Number['Conservation Reserves']

plt.figure(figsize=(10,5))
plt.plot( x, y1, marker='o', markerfacecolor='teal', color='teal',label='National Parks')
plt.plot( x, y2, marker='o', markerfacecolor='turquoise', color='turquoise',label='Wild Life Sanctuaries')
plt.plot( x, y3, marker='o', markerfacecolor='red', color='red', label='Community Reserves')
plt.plot( x, y4, marker='o', markerfacecolor='lime', color='lime', label='Conservation Reserves')

plt.style.use("seaborn-darkgrid")
plt.title('TOTAL NUMBER OF PROTECTED AREAS', fontsize = 16)
plt.xlabel('YEAR',fontsize = 14)
plt.xticks(x,fontsize = 12)
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.ylabel('COUNT',fontsize = 14)
plt.legend(bbox_to_anchor=(1.4, 1), fancybox=True, fontsize= 14)

plt.savefig('Total_no_of_Protected_Areas_India.png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close()
```

*Interpretation : There has been a rapid increase in the number of Community Reserves accross the country varying from 46 in 2018 to 214 in 2020. For the Conservation Reserves a small increase in each year is observed whereas National Parks remain more or less contant in number. The number of Wild Life sanctuaries experienced an increase from 512 in the year 2009 to 566 in the year 2019.*

```python
Area = df[['Year', 'Area Under National Parks (km2)','Area Under Wild Life Sanctuaries (km2)','Area Under Community Reserves (km2)','Area Under Conservation Reserves (km2)','Total Area under Protected Areas (km2)']]
Area.rename({'Area Under National Parks (km2)' :'National Parks', 'Area Under Wild Life Sanctuaries (km2)': 'Wild Life Sanctuaries','Area Under Community Reserves (km2)':'Community Reserves','Area Under Conservation Reserves (km2)':'Conservation Reserves','Total Area under Protected Areas (km2)':'Total Protected Area'}, axis= 1, inplace = True)
Area
```

```python
x = Area['Year']
y1 = Area['National Parks']
y2 = Area['Wild Life Sanctuaries']
y3 = Area['Community Reserves']
y4 = Area['Conservation Reserves']

plt.figure(figsize=(12,5))
plt.plot( x, y1, marker='o', markerfacecolor='teal', color='teal',label='National Parks')
plt.plot( x, y2, marker='o', markerfacecolor='turquoise', color='turquoise',label='Wild Life Sanctuaries')
plt.plot( x, y3, marker='o', markerfacecolor='red', color='red', label='Community Reserves')
plt.plot( x, y4, marker='o', markerfacecolor='lime', color='lime', label='Conservation Reserves')

plt.style.use("seaborn-whitegrid")
plt.title('TOTAL PROTECTED AREA IN $KM^{2}$', fontsize = 16)
plt.xlabel('YEAR',fontsize = 14)
plt.xticks(x,fontsize = 12)
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.ylabel('PROTECTED AREA IN $KM^{2}$',fontsize = 14)
plt.legend(bbox_to_anchor=(1.34, 1), fancybox=True, fontsize= 12)

plt.savefig('Total_Protected_Area_India.png', transparent=False, bbox_inches='tight', pad_inches=0, dpi=2000)
plt.show()
plt.close()
```

*Interpretation : National Parks observed an increase of 10% in its area from 2009 to 2020. Following the same timeline Wild Life Sanctuaries had an increase of 8% and Conservation Reserves witnessed that of 256%. The highest increase was obsereved in the Community Reserves of India with 6100% of its geographical region.*
