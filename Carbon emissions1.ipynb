```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Bruteforce004/Cleaned-data-of-Indian-Forests/main/Air%20and%20GHG%20emissionsCarbon%20dioxide%20(CO2)2009_2018.csv")
```

```python
df.head()
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
df.rename({"year co2 is measured":"Year"},axis=1,inplace=True)
df.rename({"value(Tonnes/capita)":"$CO_{2}$(Tonnes/Capita)"},axis=1,inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,8))                                  
plt.style.use("seaborn")
y = df['$CO_{2}$(Tonnes/Capita)']
x = df['Year']
width = 0.5
plt.bar(x, y, width, color = 'c')
plt.plot(x, y, color = 'black', marker = 'o')
plt.xticks(x,rotation=45, ha="center", fontsize=14)
plt.yticks(fontsize = 14)
plt.ticklabel_format(axis="y", style='plain')
plt.title("CARBON EMISSIONS(2009-18)", fontsize=20)
plt.xlabel("YEAR", fontsize = 16)
plt.ylabel("$CO_{2}$(TONNES/CAPITA)", fontsize = 16)
plt.show()
plt.close()
```

INTERPRETATION:
From the year 2008-14, emission rate was found to be increasing.
Data was found to be consistent for the later years followed by a slight increase in the year 2017-18.
